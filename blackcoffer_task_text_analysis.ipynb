```python
def extract(url):
    page = requests.get(url)
    if page.status_code != 200:
        return None, None
    soup = BeautifulSoup(page.text, 'html.parser')

    # Adjust these selectors based on the website's structure
    title = soup.find('h1').text.strip() if soup.find('h1') else None
    content = [i.text.strip() for i in soup.find_all('p')]  # Adjust as needed for specific article tags

    return title, content
```

```python
import syllapy  # You might need to install this library

def complex_word_percentage(text):
    words = nltk.word_tokenize(text)
    complex_words = [word for word in words if syllapy.count(word) > 2]
    complex_word_count = len(complex_words)
    total_words = len(words)

    percentage = (complex_word_count / total_words) * 100 if total_words > 0 else 0
    return percentage, complex_word_count
```

```python
def avg_words_per_sentence(text):
    sentences = nltk.sent_tokenize(text)
    words = nltk.word_tokenize(text)
    return len(words) / len(sentences) if sentences else 0
```

```python
def syllable_count(text):
    words = nltk.word_tokenize(text)
    syllable_counts = [syllapy.count(word) for word in words]
    return sum(syllable_counts)
```

```python
def personal_pronoun_count(text):
    personal_pronouns = ['I', 'me', 'my', 'mine', 'you', 'your', 'yours',
                         'he', 'him', 'his', 'she', 'her', 'hers',
                         'we', 'us', 'our', 'ours', 'they', 'them', 'their', 'theirs']
    words = nltk.word_tokenize(text)
    return sum(1 for word in words if word in personal_pronouns)
```
